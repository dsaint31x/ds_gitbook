```
from sklearn.datasets import load_digits
import numpy as np

digits = load_digits()
```

```
samples = [0,10,20,30]
m = list()
v = list()
for idx in range(len(samples)):
  v.append(digits.data[samples[idx]])
  m.append(digits.data[samples[idx]].reshape(8,8))
print(v[0].shape)


import matplotlib.pyplot as plt
plt.figure(figsize=(4,2))

for idx in range(len(samples)):
  plt.subplot(1,len(samples),idx+1)
  plt.imshow(m[idx], interpolation='nearest', cmap=plt.cm.bone_r)
  plt.grid(False)
  plt.xticks([]); plt.yticks([])
  plt.title(f'image {idx+1:d}')
plt.suptitle('MNIST data')
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(7,3))

for idx in range(len(samples)):
  plt.subplot(1,len(samples),idx+1)
  plt.imshow(v[idx][:,np.newaxis], aspect=0.1, interpolation='nearest', cmap=plt.cm.bone_r)
  plt.grid(False)
  plt.xticks([]); plt.xticks([])
  plt.title(f'vector {idx+1:d}')
plt.suptitle('MNIST data')
plt.tight_layout()
plt.show()
```

```
A = np.array([[11,12,13],
              [21,22,23]])
I = np.eye(2,2)
print(A.shape)
print(I.shape)
```

```
res = I@A
```

```
print( 'A and res are identical.' if np.allclose(A,res) else 'A and res are not identical.')
```
